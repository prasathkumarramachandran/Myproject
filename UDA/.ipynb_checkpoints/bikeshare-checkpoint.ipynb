```python
#Import Pandas package
import pandas as pd

#Import Numpy package
import numpy as np

#Map the city and data file names
CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }

#Define the cities, Months and Days 
in_cities = ['chicago','new york city','washington']
in_months = ['january', 'february', 'march', 'april', 'may','june','all']
in_days =   ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday','sunday','all']


#Get the city for which the analyis to be performed and Validate the name...
def get_input_city():
    city_in = str.lower(input('Would you like to see data for chicago, new york city, or washington?'))
    while city_in not in in_cities:
        print('You have entered an incorrect city name as ',city_in,'\n')   
        city_in = str.lower(input('Please choose from one of the following cities..chicago, new york city, or washington'))
    return city_in

#Get the month as input and Validate the month entered in the list of months..
def get_input_month():
        
    in_mon = str.lower(input('Which month - January, February, March, April, May, June or all?'))
    while in_mon not in in_months:        
        print('You have entered an incorrect month as',in_mon,'\n')
        in_mon = str.lower(input('Could you please choose from any of the following months ? January, February, March, April, May, June or all'))
    return in_mon
 
#Get the day entered and validate..                            
def get_input_day():    
    in_day = str.lower(input('Which day - Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday or all'))
 
 
    while in_day not in in_days:
        print('You have entered an incorrect day as',in_day,'Please choose from the below list \n')
        in_day = str.lower(input('Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday or all'))

    return in_day
    
#Let's load the data and perform the analysis and print the cool statistics...
#Data Preparation based on the user input...
def load_data(city, month, day):
 # load data file into a dataframe
    df = pd.read_csv(CITY_DATA[city])
    
 # convert the Start Time column to datetime
    df['Start Time'] = pd.to_datetime(df['Start Time'])

    
 # extract month and day of week from Start Time to create new columns
    df['month'] = df['Start Time'].dt.month
    df['day_of_week'] = df['Start Time'].dt.weekday_name
    df['hour'] = df['Start Time'].dt.hour
    df['Start_End'] = df['Start Station']+df['End Station']

 # filter by month if applicable
    if month != 'all':
 # use the index of the months list to get the corresponding int
        months = ['january', 'february', 'march', 'april', 'may', 'june']
        month = months.index(month)+1
    
 # filter by month to create the new dataframe
        df = df[df['month'] == month]

 # filter by day of week if applicable
    if day != 'all':
        df = df[df['day_of_week'] == day.title()] 
        
    return df

#Time Statistics
def time_stats(df):

 # most common month
    df_popular_month = df['month'].mode()
    print('Most Frequent Start Month:\n', in_months[int(df_popular_month-1)])

 # most common day of week
    df_popular_day =  df['day_of_week'].mode()
    print('Most Frequent Start Week Day:\n', df_popular_day)

 # most common hour of day    
    df_popular_hour = df['hour'].mode()
    print('Most Frequent Start Hour:\n', df_popular_hour)

    return df

# Station Statistics
def station_stats (df):

# most common start station
    df_popular_st_sta = df['Start Station'].mode()
    print('Most Popular Start Station:\n', df_popular_st_sta)
    
# most common end station
    df_popular_ed_sta = df['End Station'].mode()
    print('Most Popular End Station:\n', df_popular_ed_sta)

# most common trip from start to end (i.e., most frequent combination of start station and end station)
    df_popular_stend_sta = df['Start_End'].mode()
    print('Most Popular Start End Station:\n', df_popular_stend_sta)
    
    return df

#Trip Duration Stats
def trip_duration_stats(df):
    
 # Trip duration total travel time
    df_total_travel_time = df['Trip Duration'].sum()
    print('Trip duration total travel time : ',df_total_travel_time)
    
 # Trip duration average travel time
    df_average_travel_time = df['Trip Duration'].mean()
    print('Trip duration average travel time : ',df_average_travel_time)
    
    return df

#4 User info
def user_stats(df,city):
    
 # counts of each user type
  
    df_user_types = df['User Type'].value_counts()
    print('Different User Types : ', df_user_types)

 # counts of each gender (only available for NYC and Chicago)

    if city !='washington':
        
 # earliest, most recent, most common year of birth (only available for NYC and Chicago) The Files

        df_ear_birth  = df['Birth Year'].min()
        print('Earliest year of Birth : \n', df_ear_birth)
        
        df_max_birth  = df['Birth Year'].max()
        print('most recent year of Birth : \n',df_max_birth )
        
        df_common_birth     = df['Birth Year'].mode()
        print('most common year of birth : \n', df_common_birth )
        
        df_gender     = df['Gender'].value_counts()
        print('Different Gender Types : \n', df_gender)
        
 # Would like to see the data for the entered city, month and day ? 
    print ('Would like to see the 5 lines of raw data....')

    detail_data = str.lower(input('Please enter Yes or No \n \n'))

    if detail_data == 'yes':
        print(df.head(5))

        print ('\n \n Thank you, Please execute the program again,  if you wish to perform another analysis\n \n')

    if detail_data == 'no':
        print ('\n Thank you, Please execute the program again  if you wish to perform another analysis! ')
    return df


#Perform the bikes share analysis for the given user inputs and display statistics....
def bikeshare_analysis():
    
 # Get the user choices of city, month and day
    city_in = get_input_city()
    mon_in  = get_input_month()
    day_in  = get_input_day()

 #load the data from the files and prepare the data for analysis
    af = load_data(city_in,mon_in,day_in)

    print('\n Cool ! Intresting Statistics for the city {},month {},day {} is getting prepared now....\n'.format(city_in, mon_in, day_in))
    
 #Find the time statistics
    af = time_stats(af)

 #Find the station statistics
    af = station_stats (af)
    
 #Find the trip duration statistics 
    af = trip_duration_stats (af)

 #Find the user statistics
    af = user_stats(af,city_in)

    return

#Perform the bikes share analysis for the given user inputs and display statistics....
bikeshare_analysis()
    
```
