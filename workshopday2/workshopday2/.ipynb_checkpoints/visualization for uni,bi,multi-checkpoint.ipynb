# Data Visualization
Data Visualization helps get insights from Data

```python
#imporing libraries for visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
import numpy as np
```

```python
#reading the dataset and assigning it to variable "data"
data=pd.read_csv("C:/Users/jayanth/Downloads/KAG_Percent_of_Adult_diabetes__by_race_ethnicity_and_year.csv")

#total population in each demografic...
tot=data["Estimated Number of Adults Had Diabetes"]/data["Percent of Adults with Diabetes"]*100
total=[]
for i in tot:
    a=round(i,0)
    total.append(a)
total_pop=pd.DataFrame(total)
data["total_pop"]=total_pop # created a new column total population
data.head()# It shows the 5 data points or rows
```

# Uni variate visualization

```python
#histogram on continuous variable
data.hist(column="Percent of Adults with Diabetes",        # Column to plot
              figsize=(8,8),         # Plot size
              color="blue",          # Plot color
              bins=50)               # Use 50 bins 
```

#The above histogram shows the number of repeatations of "percentage of adults with diabetes".6.2-6.4 have repeated 5 times. 

```python
#Box plot
data.boxplot(column="Estimated Number of Adults Had Diabetes",        # Column to plot
                 by= "Demographic",         # Column to split upon
                 figsize= (8,8),rot=270,#to rotate the x-axis ticks to certain angle
                  grid=True)        # Figure size
```

#The above boxplot shows the distribution of population with diabetics of each demographic in all the given years 

```python
#density graph
data["Percent of Adults with Diabetes"].plot(kind="density",  # Create density plot
                      figsize=(8,8))   # Set figure size
```

```python
#distplot
sea.distplot(data["Percent of Adults with Diabetes"],#continous varible to be viewed
             hist=True,#if true histogram get displayed
             kde=True,#if true density line/graph get displayed 
             bins=int(10),#number of bins
             color = 'blue',#color of histogram and kde 
             hist_kws={'edgecolor':'black'},#displyes edge colors of the histogram
             kde_kws={'linewidth': 2.5})#width of the density line
```

```python
#bar graph for single variable
data_table=pd.crosstab(index=data["Percent of Adults with Diabetes"],columns="count")
data_table.plot(kind="bar",figsize=(8,8))
```

#we can see, how many times each percentage is repeating 

```python
#stacked bar chat
data_table = pd.crosstab(index=data["Percent of Adults with Diabetes"], 
                          columns=data["Demographic"])
 
data_table.plot(kind="bar",stacked=True,figsize=(8,8))
```

#"Black, non-Hispanic Connecticut Adults (18+y)"   has highest percentages of adults having diabetics

```python
data_table = pd.crosstab(index=data["Percent of Adults with Diabetes"], 
                          columns=data["Demographic"])

data_table.plot(kind="bar",stacked=False,figsize=(8,8))
```

#when stacked is false, then we can see repeatation of certain number for a categorical variable in seperate bars

# Bi variate visualization

```python
#Bar charts(categorical-continous)
data_2003=data.loc[(data["Data Years"]=="2003-2005")]
sea.barplot(y="Demographic",x="Percent of Adults with Diabetes",data=data_2003)
```

```python
#line graph(categorical-continous)
data_hisp=data.loc[(data["Demographic"]=="Hispanic Connecticut Adults (18+y)"),:] 
plt.plot(data_hisp["Data Years"],data_hisp["total_pop"])
plt.xticks(data_hisp["Data Years"],rotation="vertical")
```

```python
#pi-chart
data_2003_conn=data.loc[(data["Demographic"]=="Connecticut adults (18+y)")&(data["Data Years"]=="2003-2005")]
# Data to plot
labels=data_2003_conn["Estimated Number of Adults Had Diabetes"],data_2003_conn["total_pop"]-data_2003_conn["Estimated Number of Adults Had Diabetes"]
#sizes = [215, 130, 245, 210]
#colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

 
# Plot
plt.pie(labels=labels)
 
plt.axis('equal')
plt.show()
```

```python
#scatter plot
plt.scatter(x=data["Estimated Number of Adults Had Diabetes"],y=data["Percent of Adults with Diabetes"])
```

```python
plt.scatter(x=data["Data Years"],y=data["Percent of Adults with Diabetes"])
plt.xticks(data["Data Years"],rotation="vertical")
```
