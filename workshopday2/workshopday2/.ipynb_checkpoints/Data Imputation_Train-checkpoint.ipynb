# Data Imputation using Mean and Mode

```python
train=pd.read_csv("Train_UWu5bXk.csv")
train.head()


# In[49]:


mode=train["Outlet_Size"].mode()
mode[0]
```

```python
mean=round(train["Item_Weight"].mean(),1)
mean

```

```python
mode1=train["Outlet_Establishment_Year"].mode()
mode1[0]
```

```python
train["Outlet_Size"]=train["Outlet_Size"].fillna("mode[0]")
train["Outlet_Establishment_Year"]=train["Outlet_Establishment_Year"].fillna(mode1)
# mean1=round(df["Item_Weight"].mean(),1)
train["Item_Weight"]=train["Item_Weight"].fillna(mean)

```

```python
# KNN Imputation using Boston data(numerical data)
data=pd.read_csv("boston_csv.csv")
data.head()
```
