```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Train_UWu5bXk.csv")
```

```python
# IMpute values
data.isnull().sum()
data["Outlet_Size"]=data["Outlet_Size"].fillna("Medium")
data["Outlet_Establishment_Year"]=data["Outlet_Establishment_Year"].fillna(1985)
mean1=round(data["Item_Weight"].mean(),1)
data["Item_Weight"]=data["Item_Weight"].fillna(mean1)
data.head()
```

```python
data1=data.iloc[:,0:6]
data1.head()
```

```python
data2=data.iloc[:,6:12]
Item=data1.iloc[:,0]
data2["Item_Identifier"]=Item
```

```python
data2.head()
```

```python
# Selecting Item_Type where Item_Fat_content is Regular
data[data.Item_Fat_Content=="Regular"].Item_Type
```

```python
# Selecting distinct values in Item_Type
data1.Item_Type.unique()
```

## Select with multiple conditions

```python
# Select Outlet_Size=High from the data1 and select Item_Type=Dairy from the data2
data[(data2.Outlet_Size == 'High') & (data1.Item_Type == 'Dairy')]
```

```python
# selecting Item_MRP,Item_weight where Item_Fat_Content=="Low Fat",Item_Type=="Dairy"
data1[(data1.Item_Fat_Content=="Low Fat") &(data1.Item_Type=="Dairy")][["Item_MRP","Item_Weight"]]
```

## ORDER BY

```python
# selecting*from data1 where item_fat_content=Low Fat order(sorting) by MRP rate 
ascending=data1[data1.Item_Fat_Content=="Low Fat"].sort_values('Item_MRP')
ascending
```

```python
# sorting a specific column
data1.sort_values('Item_MRP')
```

```python
#select * from data1 where  Item_Fat_Content ==Regular order by Item_MRP in descending order
dscending=data1[data1.Item_Fat_Content == 'Regular'].sort_values('Item_MRP',ascending=False)
dscending
```

## IN, NOTIN

```python
# selecting a 'Soft Drinks','Meat','Household' in the Item_Type column
data_in=data1[data1.Item_Type.isin(['Soft Drinks','Meat','Household'])]
data_in
```

```python
data_not_in=data[~data.Item_Type.isin(['Soft Drinks','Meat','Household'])]
data_not_in
```

## Group by

```python
data2.groupby(["Outlet_Identifier", "Outlet_Size"]).size().to_frame('size').reset_index().sort_values(
    ["Outlet_Identifier", "Outlet_Size"])
```

```python
data2.groupby(["Outlet_Establishment_Year","Outlet_Size"]).size().to_frame('size').reset_index().sort_values(
["Outlet_Establishment_Year","Outlet_Size"],ascending=[False,False])
```

```python
#select Item_Fat_Content, count(*) from data1 where Item_Type = 'Dairy' group by Item_Fat_Content having count(*) > 100 order by count(*) desc
data1[data1.Item_Type == 'Dairy'].groupby('Item_Fat_Content').filter(lambda g: len(g) > 100).groupby('Item_Fat_Content').size().sort_values(ascending=False)
```

```python
data1[data1.Item_Type=='Soft Drinks'].groupby('Item_Type').filter(lambda g: len(g)>100).groupby('Item_Fat_Content').size().sort_values(ascending=False)
```

```python
#select Item_MRP from data1 order by size desc limit 10
data1.nlargest(10, columns='Item_MRP')
```

```python
#select max(Item_MRP), min(Item_MRP), mean(Item_MRP), median(Item_MRP) from data
data.agg({'Item_MRP': ['min', 'max', 'mean', 'median']})
```
