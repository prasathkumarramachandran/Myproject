Group By aggregations

```python
# Grouping by one or more variable(s) and aggregating one column
data.groupby('Item_Type',as_index=False)['Item_MRP'].mean()
```

```python
# Grouping by one or more variable(s) and aggregating multiple columns in same way
#Finding average Item_MRP,Item_Outlet_Sales  across Item_Type
data.groupby('Item_Type',as_index=False)[['Item_MRP','Item_Outlet_Sales']].mean()
```

```python
# Grouping by one or more variable(s) and aggregating multiple columns differently
data.groupby('Item_Type',as_index=False).agg({'Item_MRP':np.sum,'Item_Outlet_Sales':np.mean}).rename(columns={'Item_MRP':'Mean_MRP','Item_Outlet_Sales':'Total_Outlet_sales'})
```

```python

data.groupby(['Item_Type','Item_Weight'],as_index=False).agg({'Item_MRP':np.sum,'Item_Outlet_Sales':np.mean}).rename(columns={'Item_MRP':'Mean_MRP','Item_Outlet_Sales':'Total_Outlet_sales'})
```

```python
###Window functions using transform
#Sorting data within group
data['Item_Weight']=data.groupby('Item_Type')['Item_Weight'].transform(lambda x:x.sort_values(ascending=False))
```
