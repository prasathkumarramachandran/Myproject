## Missing Value Ratio

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read csv file
data = pd.read_csv("KAG_Percent_of_Adult_diabetes__by_race_ethnicity_and_year.csv")
data
```

```python
# check the percentage of missing value in each variable
data.isnull().sum()/len(data)*100
```

```python
data.shape
```

```python
data['95% Confidence Limits for Percent'].value_counts()
```

```python
# saving missing value in a variable
a = data.isnull().sum()/len(data)*100
# saving column name in a variable
variables = data.columns
variable = []
for i in range(0,7):
    # setting the threshold as 20%
    if a[i]<=20:
        variable.append(variables[i])
```

```python
variable
```

## Low Variance filter

```python
mean = round(data['Estimated Number of Adults Had Diabetes'].mean(),1)
data['Estimated Number of Adults Had Diabetes']=data['Estimated Number of Adults Had Diabetes'].fillna(mean)
data['95% Confidence Limits for Percent']=data['95% Confidence Limits for Percent'].fillna("5.9-6.7")
data
```

```python
data.isnull().sum()
```

```python
data.var()
```

```python
numeric = data[['Estimated Number of Adults Had Diabetes', 'Standard Error for Percent', 'Percent of Adults with Diabetes']]
var = numeric.var()
numeric = numeric.columns
variable = []
for i in range(0, len(var)):
    if var[i]>=10:
        variable.append(numeric[i])
```

```python
variable
```
